# This is the first section of the assignment

```python
#This section of the notebook will read the table from Wikipedia 
#Wikipedia link - https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M
#Import all required libraries
import pandas as pd
from pandas.io.html import read_html
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

wikitable = pd.read_html(page,  attrs={"class":"wikitable"}) #this will extract the tables in the web page as lists
#since we need to take the first tabel from the webepage, we are using 0. in this example, there is only one table.....
#..if there are multiple tables use the numbers accrodnigly
x=wikitable[0] 
#As mentioned in the assignment drop the row is the bourgh is not assigned
x.drop(x[x['Borough']=='Not assigned'].index, inplace=True)
import numpy as np
#As mentioned in the assignment is the neighborhood is not assigned, assign the value in borough to neighborhood
x['Neighbourhood'] = np.where(x['Neighbourhood'] == 'Not assigned', x['Borough'], x['Neighbourhood'])
#As mentioned in the assignment, group the neighborhood value based on Postcode and Borough.
x=x.groupby(['Postcode','Borough']).apply(lambda group: ', '.join(group['Neighbourhood'])).reset_index()
#renaming the column 0 to Neighborhood post the groupby function
x.rename(columns = {0:'Neighborhood'}, inplace = True) 
x
print(x.shape)
```
