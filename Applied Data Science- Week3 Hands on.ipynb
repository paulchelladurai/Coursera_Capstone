```python
#This section of the notebook will read the table from Wikipedia 
#Wikipedia link - https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M
#Import all required libraries
import pandas as pd
from pandas.io.html import read_html
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

wikitable = pd.read_html(page,  attrs={"class":"wikitable"})
x=wikitable[0]
#print(x)
x.drop(x[x['Borough']=='Not assigned'].index, inplace=True)
import numpy as np
x['Neighbourhood'] = np.where(x['Neighbourhood'] == 'Not assigned', x['Borough'], x['Neighbourhood'])
x
```

```python
x=x.groupby(['Postcode','Borough']).apply(lambda group: ', '.join(group['Neighbourhood'])).reset_index()
```
